```python
import numpy as np 
import cv2  
import tensorflow_hub as hub 
import matplotlib.pyplot as plt 
import tensorflow as tf
```

# LOAD MODEL

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

# PREPROCESSING

```python
def load_image(img_path): 
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels = 3) 
    img = tf.image.convert_image_dtype(img , tf.float32) 
    img = img[tf.newaxis,:] 
    return img 
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# LOAD IMAGES 

```python
content_img = load_image('/content/drive/MyDrive/manzara.jpeg') 
style_img = load_image('/content/drive/MyDrive/monet.jpeg')
```

## CONTENT IMAGE

```python
plt.imshow(np.squeeze(content_img))
plt.show()
```

## STYLE IMAGE

```python
plt.imshow(np.squeeze(style_img))
plt.show()
```

# FINAL OUTPUT

```python
final = model(tf.constant(content_img), tf.constant(style_img))
```

```python
plt.imshow(np.squeeze(final)) 
plt.show()
```
